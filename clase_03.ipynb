```python
import pymannkendall as mk
```

```python
mk_O3 = mk.original_test(df['O3'], alpha=0.05)
```

```python
mk_O3
```

```python
a = mk_O3.slope
b = mk_O3.intercept
```

```python
def f5(x):
    return x*a+b
```

```python
x = range(0, 3073)
fig, ax = plt.subplots(figsize=(10,4),dpi=200)
ax.plot(x, df['O3'], lw=0.3, color='black');
ax.plot(x, [f5(i) for i in x], color='red', lw=1.5)
```

```python
mk_O3 = mk.original_test(df['CO'], alpha=0.05)
a = mk_O3.slope
b = mk_O3.intercept
def f5(x):
    return x*a+b
```

```python
x = range(0, 3073)
fig, ax = plt.subplots(figsize=(10,4),dpi=200)
ax.plot(x, df['CO'], lw=0.3, color='black');
ax.plot(x, [f5(i) for i in x], color='red', lw=1.5)
```
